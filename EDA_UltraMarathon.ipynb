```python
import pandas as pd
import seaborn as sns
```

```python
race_data = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
race_data.shape
```

```python
race_data.head(20)
```

```python
race_data.info()
```

```python
# 1. Fetching only the 50km and 50m races conducted in USA in 2020
```

```python
#1a.combining the 50km and 59mi races
race_data[(race_data['Event distance/length'] == '50km') | (race_data['Event distance/length'] == '50mi')] #race_data[(race_data['Event distance/length'].isin(['50km','50mi']) works well too
```

```python
#1b. Fetching the events conducted in USA
race_data[race_data['Event name'].str.contains('USA')]
```

```python
#1c. Combining all above
race_data_USA = (race_data[(race_data['Event distance/length'].isin(['50km','50mi']))  & (race_data['Event name'].str.contains('USA')) & (race_data['Year of event'] == 2020)] ) 
```

```python
race_data_USA.shape
```

```python
race_data_USA.head(2)
```

```python
#2. calculate the athlete age
race_data_USA['athlete_age']= 2020 - race_data_USA['Athlete year of birth']
```

```python
race_data_USA.head(2)
```

```python
#3. Removing 'h' from athlete performance
race_data_USA['Athlete performance'] = race_data_USA['Athlete performance'].str.split(' ').str.get(0)
```

```python
race_data_USA.head(2)
```

```python
#4.Removing country name from event name
race_data_USA['Event name'] = race_data_USA['Event name'].str.split('(').str.get(0)
```

```python
race_data_USA.head(2)
```

```python
# checking the duplicates or null or inf values
race_data_USA.isna().sum()
```

```python
#dropping the unnecessary columns
df = race_data_USA.drop(['Athlete club','Year of event','Athlete country'],axis=1)
```

```python
df.head(3)
```

```python
df.reset_index()
```

```python
df.isna().sum()
```

```python

#dropping the unnecessary columns
df = df.drop(['Athlete year of birth','Athlete age category'],axis=1)
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
#Checking the null values
df[df['athlete_age'].isna()==1]
```

```python
df = df.dropna()
```

```python
#checking for duplicates
df[df.duplicated()]
```

```python
df.shape
```

```python
# renaming the column names as column name contains space

df = df.rename(columns={'Event dates'	: 'event_day',
                   'Event name' : 'event_name',
                   'Event distance/length' : 'event_distance',
                   'Event number of finishers':'event_number_of_finishers',
                   'Athlete performance':'athlete_performance',
                   'Athlete gender' : 'athlete_gender',
                   'Athlete average speed':'athlete_average_speed',
                    'Athlete ID':'athlete_ID'})
```

```python
# restting the index

df.reset_index(drop=True)
```

```python
df.head(10)
```

```python
df.info()
```

```python
# fixing the data types
df['athlete_age'] = df['athlete_age'].astype(int)
df['athlete_average_speed'] = df['athlete_average_speed'].astype(float)
```

```python
df.info()
```

```python
sns.barplot(data=df,x='event_distance',y='event_number_of_finishers',hue='athlete_gender')
```

```python

sns.histplot(data=df,x='event_distance',hue='athlete_gender')
```

```python
sns.violinplot(data=df,x='event_distance',y='athlete_average_speed',hue='athlete_gender')
```

```python
sns.lmplot(data=df,x='athlete_age',y='athlete_average_speed',hue='athlete_gender')
```

```python
#difference in speed for the 50km,50mi male to female

df.groupby(['event_distance','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are best in 50km,50mi race

df.groupby(['athlete_age'])['athlete_average_speed'].agg('mean','median')
```

```python
#what age groups are best in 50km race

df.query('event_distance == "50km"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

```python
df.head(10)
```

```python
#seasons for the data -> slower in summer than winter
#spring season from (march)3-5(may)
#summer 6-8
#fall 9-11
#winter 12-2
```

```python
#splitting the month
df['race_month'] = df['event_day'].str.split('.').str.get(1).astype('int')
```

```python
df['race_season'] = df['race_month'].apply(lambda x : 'Winter' if x in [12,1,2] else 'Summer' if x in [6,7,8]  else 'Spring' if x in [3,4,5] else 'Fall') 
```

```python
df.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

```python
#Considering only 50mi in above scenario
```

```python
df.query('event_distance == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
